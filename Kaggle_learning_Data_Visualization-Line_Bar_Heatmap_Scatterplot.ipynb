```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fifa_data = pd.read_csv('Data_Visualization_Data_1.csv', index_col="Date", parse_dates=True)
```

```python
fifa_data
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Add title
plt.title("FIFA World Rankings")

# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data)
```

```python
list(fifa_data.columns)
```

```python
plt.figure(figsize=(14,6))

plt.title("Germany vs Spain")

# Line chart showing fifa ranking of 'Germany'
sns.lineplot(data=fifa_data['GER'], label="Germany")

# Line chart showing fifa ranking of 'Spain'
sns.lineplot(data=fifa_data['ESP'], label="Spain")

# Add label for horizontal axis
plt.xlabel("Date")
```

```python
fifa_data_new=fifa_data.tail(10)
```

```python
fifa_data_new
```

```python
plt.figure(figsize=(14,6))
plt.title("Germany World Rankings over the years")

# Bar chart 
sns.barplot(x=fifa_data_new.index, y=fifa_data_new['GER'])

# Add label for vertical axis
plt.ylabel("World Rankings")
```

```python
plt.figure(figsize=(14,7))
plt.title("World Ranking by Date")

# Heatmap 
sns.heatmap(data=fifa_data_new, annot=True)

#annot=True - This ensures that the values for each cell appear on the chart. 
#(Leaving this out removes the numbers from each of the cells!)

plt.xlabel("Country")
```

```python
insurance_data = pd.read_csv('Data_Visualization_Data_2.csv')
```

```python
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
#To double-check the strength of the relationship, you might like to add a regression line, or the line that best fits the data. 
#We do this by changing the command to sns.regplot.

sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
#We can use scatter plots to display the relationships between (not two, but...) three variables! 
#One way of doing this is by color-coding the points.

sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
#This scatter plot shows that while nonsmokers to tend to pay slightly more with increasing BMI, smokers pay MUCH more.
#To further emphasize this fact, we can use the sns.lmplot command to add two regression lines

sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

```python
# categorical scatter plot

sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```
